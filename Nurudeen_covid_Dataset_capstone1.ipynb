```python
covid_df = pd.get_dummies(covid_df, drop_first = True)
```

```python
covid_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype == 'object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
covid_df = covid_df.apply(lambda x: object_to_int(x))
covid_df.head()
```

```python
covid_df.info()
```

# Spliting the data into X and y

```python
X = covid_df.drop(columns= ['CLASIFFICATION_FINAL'])
y = covid_df['CLASIFFICATION_FINAL']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state= 40)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train.values.ravel())
```

```python
feat_score = pd.DataFrame({'Fraction of variables': rf.feature_importances_}, index = X.columns)
feat_score = feat_score.sort_values(by = "Fraction of variables")
feat_score.plot(kind = 'barh', figsize = (0.5, 50))
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
```

```python
model_lr.fit(X_train, y_train)
```

```python
y_predict = model_lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True)
```

```python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.svm import LinearSVC
```

```python
model_svc = LinearSVC(max_iter=100000)
model_svm = CalibratedClassifierCV(model_svc)
model_svm.fit(X_train, y_train)
```
